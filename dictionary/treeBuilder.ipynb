```python
from pprint import pp
lines = []
with open("filtered_opencorpora.txt", "r", encoding='utf-8') as f:
    line = f.readline()
    while line:
        if line != '\n' and not line[0].isdigit(): 
            lines.append(line[:-1])
        line = f.readline()

lines = lines[56168:56213]
print(len(lines))
```

```python
%%time
sep_lines = ['/'.join(['^'] + list(line) + ['$']) for line in lines]
```

```python
%%time
from bigtree import list_to_tree, print_tree

root = list_to_tree(sep_lines)
print_tree(root, style="ascii")
```

```python
%%time
from bigtree import tree_to_dict
dicted = tree_to_dict(root)
```

```python
%%time
import json

with open('huge_tree.json', 'w+', encoding='utf-8') as j:
    json.dump(dicted, j, ensure_ascii=True, indent=4)
```
